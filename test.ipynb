```python
from semanticstore import KV
from faker import Faker 
import numpy as np
```

```python
fake = Faker()
num_pairs = 1000
kv = KV('ada_lovelace_v8', num_dimensions=512)
```

```python
data_to_insert = []
for i in range(num_pairs):
    value = {
        'vector': [i] * 512,
        'payload': {
            'some': fake.word(),
            'other_data': fake.sentence()
        }
    }
    data_to_insert.append(value)
```

```python
for i, data in enumerate(data_to_insert) :
    kv[i] = data
```

```python
kv.search(query=np.random.rand(512).astype('float32'), top_k=100).filter('[].distance')
```
